```python

# Data
number = [17, 17, 31, 23, 14, 19, 22, 17, 34, 34, 12, 32, 31, 16, 38, 39, 20,
          26, 34, 11, 37, 35, 29, 34, 36, 22, 10, 14, 23, 24, 30, 18, 38, 20,
          10, 16, 11, 32, 18, 10, 26, 31, 28, 10, 23, 39, 24, 24, 22, 38]

# Bins
bin_list = [10, 15, 20, 25, 30, 35, 40]

# Plot
plt.figure(figsize=(8, 5))
plt.hist(number, bins=bin_list, color="steelblue", edgecolor="black", linewidth=1.5, alpha=0.8)

# Titles and labels
plt.title("Random Numbers Between 10 to 40", fontsize=15, fontweight="bold")
plt.xlabel("Number Range", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

# Enhancements
plt.xticks(bin_list)  # Ensure proper bin labels
plt.grid(axis="y", linestyle="--", alpha=0.6)

# Show plot
plt.show()
```
