```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
"""description: A box plot (also known as a box-and-whisker plot) in Matplotlib is a graphical representation 
that summarizes the distribution of a dataset using five key summary statistics:

1.Minimum (smallest value, excluding outliers)
2.First quartile (Q1) (25th percentile)
3.Median (Q2) (50th percentile)
4.Third quartile (Q3) (75th percentile)
5.Maximum (largest value, excluding outliers)
Additionally, box plots often show outliers(upper and lower part) as individual points beyond the "whiskers."""
```

```python
x=[43,65,76,40]

plt.boxplot(x)
plt.show()
```

```python

y=[1,2,3,4,10] # here the circle represent the extra value

plt.boxplot(y)

plt.show()
```

```python
y1=[-14,20,30,40,50]
plt.boxplot(y1)
plt.show()
```

```python



# Generate data with some outliers
np.random.seed(42)
data = np.append(np.random.randn(100) * 10 + 50, [10, 100])  # Adding two outliers

# Create a box plot
plt.boxplot(data, vert=True, patch_artist=True, flierprops=dict(marker='o', markersize=8))

plt.title("Box Plot with Outliers")
plt.ylabel("Values")
plt.show()
```

```python
# multiple boxplot 
# Generate data with some outliers
np.random.seed(42)
data2 = np.append(np.random.randn(100) * 10 + 50, [10, 100])  # Adding two outliers
data3 = np.append(np.random.randn(100)*10+70,[10,100]) 
dataset=[data2,data3]

# Create a box plot
plt.boxplot(dataset,tick_labels=["ages","weights"], vert=True, patch_artist=True, flierprops=dict(marker='o', markersize=8))


plt.title("Box Plot with Outliers")
plt.ylabel("Values")
plt.show()
```
