```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.step(x,y)

plt.xlabel("years")
plt.ylabel("price")
plt.show()
```

```python
#data
x=[1,2,3,4,5]
y=[3,5,8,4,1]

#create plot
plt.step(x,y, color="green", marker="o",ms=8,mfc="black")#mfc-> marker face color, ms-> marker color

#customization
plt.title("price sheet", fontsize=15, fontweight="bold")
plt.xlabel("years", fontsize=15)
plt.ylabel("price",fontsize=15)
plt.grid(alpha=0.8)

#show plot
plt.show()
```

```python

# Data
years = [2015, 2016, 2017, 2018, 2019, 2020]
prices = [10, 15, 25, 18, 30, 22]

# Create Step Plot
plt.figure(figsize=(8, 5))  # Adjust figure size
plt.step(years, prices, color="darkorange", marker="s", markersize=10, markerfacecolor="navy", linestyle="-.", linewidth=2)

# Customization
plt.title("Yearly Price Fluctuation", fontsize=16, fontweight="bold", color="darkred")
plt.xlabel("Years", fontsize=14, fontweight="bold")
plt.ylabel("Price ($)", fontsize=14, fontweight="bold")
plt.xticks(years)  # Ensure all years are labeled
plt.yticks(range(min(prices), max(prices) + 5, 5))  # Adjust y-ticks for better scale
plt.grid(alpha=0.6, linestyle="--")  # Light dashed grid
plt.axhline(y=min(prices), color="gray", linestyle="dotted")  # Reference horizontal line
plt.axvline(x=2015, color="gray", linestyle="dotted")  # Reference vertical line

# Show plot
plt.show()
```
