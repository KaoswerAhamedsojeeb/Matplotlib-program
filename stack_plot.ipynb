```python
import matplotlib.pyplot as plt
```

```python

# Data
x = [1, 2, 3, 4, 5, 6, 7, 8]
area1 = [13, 16, 11, 17, 15, 19, 12, 17]
area2 = [6, 8, 9, 6, 7, 6, 5, 7]
area3 = [9, 5, 1, 7, 4, 8, 2, 4]

# Labels and Colors
labels = ["Area 1", "Area 2", "Area 3"]
colors = ['skyblue', 'lightcoral', 'lightgreen']

# Create Stack Plot
plt.figure(figsize=(8, 5))  # Adjust figure size
plt.stackplot(x, area1, area2, area3, labels=labels, colors=colors, alpha=0.7)

# Customization
plt.xlabel("X Axis")
plt.ylabel("Values")
plt.title("Stacked Area Plot Example")
plt.legend(loc="upper left")  # Set a clear position for the legend
plt.grid(True, linestyle="--", alpha=0.5)  # Add grid for better readability

# Show plot
plt.show()
```

```python


# Data
x = range(1, 11)  # X-axis values from 1 to 10
category1 = [3, 6, 9, 12, 8, 10, 7, 13, 9, 11]
category2 = [5, 3, 6, 8, 6, 9, 5, 7, 6, 8]
category3 = [2, 4, 5, 7, 5, 6, 3, 6, 4, 5]

# Labels and Colors
labels = ["Category A", "Category B", "Category C"]
colors = ["royalblue", "tomato", "gold"]

# Create Stack Plot
plt.figure(figsize=(9, 5))
plt.stackplot(x, category1, category2, category3, labels=labels, colors=colors, alpha=0.75)

# Customization
plt.xlabel("Time (Days)")
plt.ylabel("Values")
plt.title("Stacked Area Chart Example")
plt.legend(loc="upper left")
plt.grid(True, linestyle="--", alpha=0.6)

# Show plot
plt.show()
```
