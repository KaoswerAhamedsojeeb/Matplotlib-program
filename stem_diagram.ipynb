```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[19,23,16,20,28,30]
y=[5,6,2,7,9,3]
z=[8,3,6,4,2,5]
```

```python
plt.stem(x,y)
plt.show()
```

```python
plt.stem(x,y,linefmt="--")
plt.stem(x,z,markerfmt='black')

plt.legend(['jan','feb'])
plt.show()
```

```python
x1 = np.linspace(0.1, 2 * np.pi, 41) 
y1 = np.exp(np.sin(x1)) 
  
plt.stem(x1, y1) 
plt.show() 

```
