```python
x=["python","c","c++","java"]
plt.xlabel("languages")
y=[80,60,70,75]
plt.ylabel("users percentage")
plt.bar(x,y, width=0.3,color=["brown","lightgreen","yellow","skyblue"],edgecolor="black",linewidth=2, label=['python','C','C++','java'])
plt.title("popular languages")
plt.legend()
plt.show()
```

```python

# Data
x_labels = ["Lenovo", "Doyel", "Dell", "HP", "Walton"]
users1 = [34, 40, 60, 53, 10]
users2 = [28, 43, 32, 45, 12]

# Bar width and positions
width = 0.3
idx = np.arange(len(x_labels))
idx2 = idx + width

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(idx, users1, width=width, label="Users 1", color='skyblue', edgecolor='black')
plt.bar(idx2, users2, width=width, label="Users 2", color='salmon', edgecolor='black')

# Labels and title
plt.xticks(idx + width / 2, x_labels, rotation=30)
plt.xlabel("Brands")
plt.ylabel("Number of Users")
plt.title("Comparison of Users by Brand")
plt.legend()

# Display the plot
plt.show()
```

```python


# Data
x_labels = ["Lenovo", "Doyel", "Dell", "HP", "Walton"]
users1 = [34, 40, 60, 53, 10]
users2 = [28, 43, 32, 45, 12]

# Bar width and positions
width = 0.3
idx = np.arange(len(x_labels))
idx2 = idx + width

# Create the plot
fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(idx, users1, width=width, label="Users 1", color='skyblue', edgecolor='black')
bars2 = ax.bar(idx2, users2, width=width, label="Users 2", color='salmon', edgecolor='black')

# Add labels above bars
for bar in bars1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height()}', 
            ha='center', va='bottom', fontsize=10, fontweight='bold')

for bar in bars2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height()}', 
            ha='center', va='bottom', fontsize=10, fontweight='bold')

# Labels and title
ax.set_xticks(idx + width / 2)
ax.set_xticklabels(x_labels, rotation=30)
ax.set_xlabel("Brands")
ax.set_ylabel("Number of Users")
ax.set_title("Comparison of Users by Brand")
ax.legend()

# Display the plot
plt.show()

```
