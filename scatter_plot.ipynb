```python
import matplotlib.pyplot as plt
```

```python


# Data
days = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
screen_time = [6, 5, 4, 7, 5, 8, 9]

# Finding peak and lowest points
max_index = screen_time.index(max(screen_time))
min_index = screen_time.index(min(screen_time))

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(days, screen_time, color="dodgerblue", s=100, edgecolor="black", label="Screen Time")

# Adding peak and lowest point annotations
plt.text(days[max_index], screen_time[max_index] + 0.3, "Peak Point", ha="center", fontsize=12, color="red", fontweight='bold')
plt.text(days[min_index], screen_time[min_index] - 0.5, "Lowest Point", ha="center", fontsize=12, color="green", fontweight='bold')

# Titles and labels
plt.title("Screen Time Per Day", fontsize=15, fontweight='bold')
plt.xlabel("Days", fontsize=12)
plt.ylabel("Hours", fontsize=12)
plt.xticks(rotation=25)  # Rotate x-axis labels for better readability
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.legend()
plt.show()
```
